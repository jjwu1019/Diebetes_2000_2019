```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
#load csv
DiabetesAtlasData_Total_Percentage_df = pd.read_csv("DiabetesAtlasData_Total_Percentage.csv")
DiabetesAtlasData_Total_Number_df = pd.read_csv("DiabetesAtlasData_Total_Number.csv")
DiabetesAtlasData_Age_Percentage_df = pd.read_csv("DiabetesAtlasData_Age_Percentage.csv")
DiabetesAtlasData_Age_Number_df = pd.read_csv("DiabetesAtlasData_Age_Number.csv")
DiabetesAtlasData_Gender_Percentage_df = pd.read_csv("DiabetesAtlasData_Gender_Percentage.csv")
DiabetesAtlasData_Gender_Number_df = pd.read_csv("DiabetesAtlasData_Gender_Number.csv")
DiabetesAtlasData_Race_Percentage_df = pd.read_csv("DiabetesAtlasData_Race_Percentage.csv")
DiabetesAtlasData_Race_Number_df = pd.read_csv("DiabetesAtlasData_Race_Number.csv")
DiabetesAtlasData_Education_Percentage_df = pd.read_csv("DiabetesAtlasData_Education_Percentage.csv")
DiabetesAtlasData_Education_Number_df = pd.read_csv("DiabetesAtlasData_Education_Number.csv")

DiabetesAtlasData_Physical_Inactivity_2018_df = pd.read_csv("DiabetesAtlasData_Physical_Inactivity_2018.csv")
DiabetesAtlasData_Obesity_2018_df = pd.read_csv("DiabetesAtlasData_Obesity_2018.csv")
DiabetesAtlasData_Newly_Diagnosed_2018_df = pd.read_csv("DiabetesAtlasData_Newly_Diagnosed_2018.csv")
DiabetesAtlasData_Diagnosed_2018_df = pd.read_csv("DiabetesAtlasData_Diagnosed_2018.csv")
```

```python
Physical_Inactivity_2018_df = DiabetesAtlasData_Physical_Inactivity_2018_df[['Year','County','State','Physical Inactivity Percentage']]
Obesity_2018_df = DiabetesAtlasData_Obesity_2018_df[['Obesity Percentage']]
Newly_Diagnosed_2018_df = DiabetesAtlasData_Newly_Diagnosed_2018_df[['Newly Diagnosed Diabetes Rate per 1000']]
Diagnosed_2018_df = DiabetesAtlasData_Diagnosed_2018_df[['Diagnosed Diabetes Percentage']]
frames = [Physical_Inactivity_2018_df, Obesity_2018_df, Newly_Diagnosed_2018_df, Diagnosed_2018_df]
Social_Determinants_2018_df = pd.concat(frames, axis=1, join="outer")

#Save df as csv file
Social_Determinants_2018_df.to_csv("Social_Determinants_combined_2018.csv", index=False, header=True)
Social_Determinants_2018_df.set_index('Year', inplace=True)
Social_Determinants_2018_df
```

```python
Total_Number_df = DiabetesAtlasData_Total_Number_df[['Year','Total - Number in 1000s']]
Age_Number_df = DiabetesAtlasData_Age_Number_df[['18-44 - Number in 1000s','45-64 - Number in 1000s','65-74 - Number in 1000s','75+ - Number in 1000s']]
Gender_Number_df = DiabetesAtlasData_Gender_Number_df[['Male - Number in 1000s','Female - Number in 1000s']]
Race_Number_df = DiabetesAtlasData_Race_Number_df[['Hispanic - Number in 1000s','Non-Hispanic White - Number in 1000s','Non-Hispanic Black - Number in 1000s','Non-Hispanic Asian - Number in 1000s']]
Education_Number_df = DiabetesAtlasData_Education_Number_df[['< High School - Number in 1000s','High School - Number in 1000s','> High School - Number in 1000s']]
frames = [Total_Number_df, Age_Number_df, Gender_Number_df, Race_Number_df, Education_Number_df]
Number_df = pd.concat(frames, axis=1, join="outer")

#Save df as csv file
Number_df.to_csv("diabetes_number_combined_2000_2019.csv", index=False, header=True)
Number_df.set_index('Year', inplace=True)
Number_df
```

```python
Total_Percentage_df = DiabetesAtlasData_Total_Percentage_df[['Year','Total - Percentage']]
Age_Percentage_df = DiabetesAtlasData_Age_Percentage_df[['18-44 - Percentage','45-64 - Percentage','65-74 - Percentage','75+ - Percentage']]
Gender_Percentage_df = DiabetesAtlasData_Gender_Percentage_df[['Male - Percentage','Female - Percentage']]
Race_Percentage_df = DiabetesAtlasData_Race_Percentage_df[['Hispanic - Percentage','Non-Hispanic White - Percentage','Non-Hispanic Black - Percentage','Non-Hispanic Asian - Percentage']]
Education_Percentage_df = DiabetesAtlasData_Education_Percentage_df[['< High School - Percentage','High School - Percentage','> High School - Percentage']]
frames = [Total_Percentage_df, Age_Percentage_df, Gender_Percentage_df, Race_Percentage_df, Education_Percentage_df]
Percentage_df = pd.concat(frames, axis=1, join="outer")

#Save df as csv file
Percentage_df.to_csv("diabetes_percentage_combined_2000_2019.csv", index=False, header=True)
Percentage_df.set_index('Year', inplace=True)
Percentage_df
```
